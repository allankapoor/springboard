```python
#create borough list from unique values in London Borough column
borough_list = list(properties_melt['London Borough'].unique())
```

```python
#create df for storing rent change values for each borough (this can't be the moest efficient way?)
pcnt_change_empty = [float(0)] * len(borough_list)
list_labels = ['Borough','Pcnt_Change']
list_cols = [borough_list, pcnt_change_empty]
zipped = list(zip(list_labels, list_cols))
data = dict(zipped)
borough_rent_change = pd.DataFrame(data) 
borough_rent_change = borough_rent_change.set_index('Borough', drop=True)
borough_rent_change
```

```python
#loop through boroughs
for borough in borough_list:
    
    #get average rent in 1998
    rent_1998 = rent_by_year.loc[(borough,1998),:]
    
    #get average rent in 2018
    rent_2018 = rent_by_year.loc[(borough,2018),:]
    
    #calculate percent change
    pcnt_change = ((rent_2018 - rent_1998)/rent_1998)*100
    
    borough_rent_change.loc[borough,:] = round(float(pcnt_change),2)
```

```python
#display output table
borough_rent_change
```

```python
#sort output table by rent increase
borough_rent_change = borough_rent_change.sort_values('Pcnt_Change',ascending=False)
borough_rent_change
```

```python
#plot bar graph of rent increases
borough_rent_change.plot.bar()
plt.xlabel('Borough')
plt.ylabel('Percent Increase')
plt.title('Increase in Rent 1998-2018 by Borough')
plt.figure(figsize=(20,20))
plt.show()
```
